```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# Nome del modello pre-addestrato utilizzato
model_name = "cardiffnlp/twitter-roberta-base-sentiment-latest"

# Caricamento del tokenizer associato al modello che prendiamo in esame
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Caricamento del modello per la classificazione delle sequenze
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
# Esempi di testo dai social media
texts = [
    "This company's performance is truly outstanding.",
    "Their service was a real letdown for me.",
    "I find the new product quite intriguing.",
    "Nothing special to report, just a standard day."
]

# Token to Text
encoded_input = tokenizer(texts, padding=True, truncation=True, return_tensors='pt')

# Lancia l'inferenza del modello
with torch.no_grad():
    output = model(**encoded_input)

# Estrai le probabilità associate ai sentiment.
probabilities = torch.nn.functional.softmax(output.logits, dim=-1)

# Recupera le etichette dei sentiment (positive, negative, neutral)
predicted_labels = torch.argmax(probabilities, dim=-1)

# Associa gli indici alle etichette
labels = ['negative', 'neutral', 'positive']
predicted_sentiments = [labels[label] for label in predicted_labels.tolist()]

# Mostra i risultati
for text, sentiment in zip(texts, predicted_sentiments):
    print(f"Testo: '{text}' - Sentiment: {sentiment}")
```
